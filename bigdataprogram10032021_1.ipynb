# Decision Tree Classifier

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('tree').getOrCreate()
df = spark.read.csv('/content/drive/MyDrive/IDS2018/02-15-2018.csv', inferSchema=True, header=True)
df.printSchema()
```

```python
from pyspark.ml.feature import VectorAssembler

assembler1 = VectorAssembler(inputCols = ["Active Mean","Fwd Header Len","Flow IAT Min","Fwd Pkt Len Mean"], outputCol = 'features')
```

```python
output = assembler1.transform(df)
```

```python
from pyspark.ml.feature import StringIndexer

indexer1 = StringIndexer(inputCol = 'FIN Flag Cnt', outputCol = 'FIN Flag CntIndex')
outputFixed1 = indexer1.fit(output).transform(output)
outputFixed1.printSchema()
```

```python
final_df1 = outputFixed1.select('features', 'FIN Flag CntIndex')
final_df1.show(3)
```

```python
train1, test1 = final_df1.randomSplit([0.7, 0.3])
```

```python
from pyspark.ml.classification import DecisionTreeClassifier
```

```python
from pyspark.ml import Pipeline
```

```python
dtree = DecisionTreeClassifier(labelCol = 'FIN Flag CntIndex', featuresCol = 'features')
```

```python
dtree_model = dtree.fit(train1)
```

```python
dtree_predictions = dtree_model.transform(test1)
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator
```

```python
binary_evaluator1 = BinaryClassificationEvaluator(labelCol = 'FIN Flag CntIndex')
```

```python
print('Decision Tree:', binary_evaluator1.evaluate(dtree_predictions))
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

multi_evaluator1 = MulticlassClassificationEvaluator(labelCol = 'FIN Flag CntIndex', metricName = 'accuracy')
print('Decision_Tree_Acc:', multi_evaluator1.evaluate(dtree_predictions))
```
