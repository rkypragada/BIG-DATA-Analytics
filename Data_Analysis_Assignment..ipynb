# Task 2: Perform_PCA_Training_Dataset

```python
import pandas as pd
import numpy as np 

DF_PCA = Training_set_I
DF_PCA.head(1000)
```

```python
from sklearn.preprocessing import StandardScaler
X_scaled = StandardScaler().fit_transform(Training_set_I)
X_scaled[:12]
Y=Training_set_I.L
```

```python
features = X_scaled.T
cov_matrix = np.cov(features)
cov_matrix[:12]
```

```python
values, vectors = np.linalg.eig(cov_matrix)
values[:12]
```

```python
vectors[:12]
```

```python
explained_variances = []
for i in range(len(values)):
    explained_variances.append(values[i] / np.sum(values))
 
print(np.sum(explained_variances), '\n', explained_variances)
```

```python
projected_1 = X_scaled.dot(vectors.T[0])
projected_2 = X_scaled.dot(vectors.T[1])
res = pd.DataFrame(projected_1, columns=['PC1'])
res['PC2'] = projected_2
res['Y'] = Y
res.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(20, 10))
sns.scatterplot(res['PC1'],[0] * len(res), hue=res['Y'], s=200)
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(res['PC2'], [0] * len(res), hue=res['Y'], s=100)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
import pandas as pd
from sklearn.preprocessing import StandardScaler
plt.style.use('ggplot')

X = Training_set_I
y = Training_set_I.L

scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

pca = PCA(n_components=2) 
X_new = pca.fit_transform(X) 
```

```python
fig, axes = plt.subplots(1,2)
axes[0].scatter(X[:,0], X[:,1], c=y)
axes[0].set_xlabel('x1')
axes[0].set_ylabel('x2')
axes[0].set_title('Without implementing PCA')
axes[1].scatter(X_new[:,0], X_new[:,1], c=y)
axes[1].set_xlabel('PC1')
axes[1].set_ylabel('PC2')
axes[1].set_title('With implementing PCA')
plt.show()
```

```python
train_x = train_data.drop(columns=['L'],axis=1)
train_y = train_data['L']
```

```python
test_x = test_data.drop(columns=['L'],axis=1)
test_y = test_data['L']
```

```python
model = LinearRegression()
model.fit(train_x,train_y)
```

```python
print('\nCoefficient of model :', model.coef_)

print('\nIntercept of model',model.intercept_)
```

```python
predict_train = model.predict(train_x)

rmse_train = mean_squared_error(train_y,predict_train)**(0.5)
print('\nRMSE on train dataset : ', rmse_train)

predict_test = model.predict(test_x) 

rmse_test = mean_squared_error(test_y,predict_test)**(0.5)
print('\nRMSE on test dataset : ', rmse_test)
```

```python
print(regressor.intercept_)

print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
df
```

```python
df1 = df.head(25)
df1.plot(kind='bar',figsize=(16,10),color=('green','yellow'))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='blue')
plt.show()
```

```python
print('mean_absolute_error:', metrics.mean_absolute_error(y_test, y_pred))  
print('mean_Squared_error:', metrics.mean_squared_error(y_test, y_pred))  
print('root_mean_squared_error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
