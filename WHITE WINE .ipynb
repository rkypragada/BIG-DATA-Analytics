# Task 2: Principal Component Analysis Implementation

# Feature Importance

```python
numeric_data = White_Wine.drop('Unnamed: 0', axis=1)
print (numeric_data.head( ))
print (numeric_data.tail( ))
numeric_data = (numeric_data - np.mean(numeric_data)) / np.std(numeric_data)

print(numeric_data.head( ))
print(numeric_data.tail( ))
```

```python
numeric_data = (numeric_data - np.mean(numeric_data)) / np.std(numeric_data)
print(numeric_data.head( ))
print(numeric_data.tail( ))
```

```python
import sklearn.decomposition
pca = sklearn.decomposition.PCA(n_components=2)
principal_components = pca.fit(numeric_data).transform(numeric_data)
```

```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib.backends.backend_pdf import PdfPages
observation_colormap = ListedColormap(['red', 'black'])
x = principal_components[:,0]
y = principal_components[:,1]
plt.figure(figsize=(10,6))
plt.scatter(White_Wine['volatile.acidity'],White_Wine['total.sulfur.dioxide'],c=White_Wine['density'],edgecolors='k',alpha=0.75,s=150)
plt.grid(True)
plt.title("P C A PLOT",fontsize=15)
plt.xlabel("Principal Component 1",fontsize=15)
plt.ylabel("Principal Component 2",fontsize=15)
plt.show()
```

# Task 3: Dividing the white_wine dataset into a training set, a validation set, and a test set.

```python
import numpy as np
import pandas as pd

#length of data 
N = 4898 
scale=5


#generated random data
X, y = np.arange(N*scale).reshape((N, scale)), np.arange(N)

#Works for pandas dataframe too

def train_test_val(X, y, train_ratio, test_ratio, val_ratio):
    assert sum([train_ratio, test_ratio, val_ratio])==1.0, "wrong given ratio, all ratios have to sum to 1.0"
    assert X.shape[0]==len(y), "X and y shape mismatch"

    ind_train = int(round(X.shape[0]*train_ratio))
    ind_test = int(round(X.shape[0]*(train_ratio+test_ratio)))

    X_train = X[:ind_train]
    X_test = X[ind_train:ind_test]
    X_val = X[ind_test:]

    y_train = y[:ind_train]
    y_test = y[ind_train:ind_test]
    y_val = y[ind_test:]

    return X_train, X_test, X_val, y_train, y_test, y_val
# put ratio as you wish
X_train, X_test, X_val, y_train, y_test, y_val=train_test_val(X, y, 0.6, 0.2, 0.2) 
```

# Task 5: Conclusion

The dataset comes from the https://archive.ics.uci.edu/ml/machine learning-databases/wine-quality/ repository of UCI machine learning databases. The original data was 4898 white wine observations. Each has the consistency of the wine and 11 chemical qualities: fixed acidity, volatile acidity, citric acid, sugar residues, chlorides, sulfur dioxide free, sulfur dioxide, sulfur dioxyde, density. The relationship between wine quality and chemical attributes has been investigated. Implemented Primary Component Analysis (PCA), a method used to reduce the data size, improve interpretability and mitigate the information loss. It achieves this by generating new variables that successively increase the variance that was implemented effectively on the dataset of white wines. Choose Machine Learning algorithm for Logistic Regression. The logistic regression explains and estimates the relationship between a binary variable dependent and an independent. The overall accuracy of 96.46 percent is achieved in this algorithm
